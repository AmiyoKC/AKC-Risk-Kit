## Random Walk Generation

```python
import numpy as np
import pandas as pd

def gbm(n_years=10,n_scenarios=1000,mu = 0.07, sigma = 0.15,steps_per_year=12, s_0=100.0):
    """
    Evolution of a stock price using a geomtrice Brownian motion model
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    rets_plus_1 = np.random.normal(loc=(1+(mu*dt)),scale=sigma*np.sqrt(dt),size=(n_steps, n_scenarios))
#    rets = pd.DataFrame(rets)
    price = s_0*pd.DataFrame(rets_plus_1).cumprod()
    return price
    
```

```python
def gbm0(n_years=10,n_scenarios=1000,mu = 0.07, sigma = 0.15,steps_per_year=12, s_0=100.0):
    """
    Evolution of a stock price using a geomtrice Brownian motion model
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    xi = np.random.normal(size=(n_steps, n_scenarios))
    rets = (mu*dt) + sigma*(dt**0.5)*xi
    rets = pd.DataFrame(rets)
    price = s_0*(1+rets).cumprod()
    return price
```
