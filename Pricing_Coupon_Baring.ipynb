```python
price_10 =ark.bond_price(10,100,0.05,12,rates)
```

```python
price_10[[1,2,3]].tail()
```

```python
price_30 =ark.bond_price(30,100,0.05,12,rates)
price_30[[1,2,3]].tail()
```

```python
rets_30 = ark.bond_total_return(price_30,100,0.05,12)
rets_10 = ark.bond_total_return(price_10,100,0.05,12)
rets_bonds = 0.6*rets_10 + 0.4*rets_30
```

```python
mean_rets_bonds = rets_bonds.mean(axis='columns')
ark.summary_stats(pd.DataFrame(mean_rets_bonds))
```

```python
price_eq = ark.gbm(n_years=10,n_scenarios=500,mu=0.07,sigma=0.15)
```

```python
rets_eq = price_eq.pct_change().dropna()
rets_eq
```

```python
rets = 0.7*rets_eq + 0.3*rets_bonds
rets_mean = rets.mean(axis='columns')
ark.summary_stats(pd.DataFrame(rets_mean))
```
